Histograms and Density Plots

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

print('Setup Complete')
```

```python
iris_data = pd.read_csv('iris.csv', index_col='Id')
```

```python
iris_data.head()
```

```python
iris_data.describe()
```

```python
plt.figure(figsize=(15,8))
plt.title('Histogram Chart for Sepal Width')
sns.histplot(iris_data['Sepal Width (cm)'])
```

```python
# Kernel Density Estimate (KDE) plots
```

```python
plt.figure(figsize=(16,8))
plt.title('Density Plot for Petal Length')
sns.kdeplot(data= iris_data['Petal Length (cm)'], shade =True)
```

```python
# 2D KDE Plots
```

```python
plt.figure(figsize=(16,8))
sns.jointplot(x= iris_data['Petal Length (cm)'], y= iris_data['Sepal Width (cm)'], kind = 'kde')
```

```python
# Histograms for each Species
```

```python
plt.figure(figsize=(10,6))
plt.title('Histogram of Petal Length by Species')
sns.histplot(data= iris_data, x= 'Petal Length (cm)', hue = 'Species')
```

```python
# KDE for each Species
```

```python
plt.figure(figsize=(10,6))
plt.title('Distribution of Petal Length by Species')
sns.kdeplot(data = iris_data, x= 'Petal Length (cm)', hue = 'Species', shade =True)
```
